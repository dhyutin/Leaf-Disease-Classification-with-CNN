```python
# Function to perform Image Segmentation

def image_segmentation(img):
    res = img
    
    # 1. Canny Edge Detection
    img = cv2.Canny(img,100,200)
    
    # 2. Gaussian Blur to smoothen the Image
    blur = cv2.GaussianBlur(img, (25,25),0)
    
    # 3. OTSU Thresholding 
    ret_otsu,im_bw_otsu = cv2.threshold(blur,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    
    # 4. Morphological operations - Closing
    kernel = np.ones((5,5),np.uint8)
    closing = cv2.morphologyEx(im_bw_otsu, cv2.MORPH_CLOSE, kernel)
    
    # 5. Sobel Filter
    sobelx64f = cv2.Sobel(closing,cv2.CV_64F,1,0,ksize=5)
    abs_sobel64f = np.absolute(sobelx64f)
    sobel_8u = np.uint8(abs_sobel64f)  
    ret_sobel,im_bw_sobel = cv2.threshold(sobel_8u,1,255,cv2.THRESH_BINARY)
    
    # 6. Morphological Operations - Closing
    kernel = np.ones((5,5),np.uint8)
    closing = cv2.morphologyEx(im_bw_otsu, cv2.MORPH_CLOSE, kernel)
    
    # 7. Image Multiplication - Multiply mask with original image
    for i in range(len(img)):
        for j in range(len(img[0])):
            if (closing[i][j]==0):
                res[i][j]=0
    return res
```
